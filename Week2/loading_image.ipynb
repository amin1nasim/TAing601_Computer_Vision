# Loading Image in Python
There are multiple ways to read or write an image in python. Each method might have different pros and cons.
In this notebook we cover three different libraries to perform ``imread()``.
- Pillow
- Matplotlib
- OpenCV

So let's get started!

## Pillow
Pillow is one of the most popular libraries to deal with images. In this course we will mostly use this library since, as we will see in future, PyTorch likes Pillow better :D.

````python
from PIL import Image
img = Image.open('cat.jpg')

display(img) # To display the image in notebook
# img.show()
````

````python
# Type
print(type(img))

# Converting to numpy
import numpy as np
img_np = np.array(img)
print(img_np.shape, img_np.dtype)

# Channels
print("Channels order: RGB")

# Converting from numpy to PIL
img = Image.fromarray(img_np) # works if numpy object is uint8

# Write the image
img.save('cat_PIL.jpg')
````

## Matplotlib
Matplotlib is able to read and write images. This library is very important scince it provide us with functions to plot more complex figures (adding axis, subplots, adding annutations, etc.)

````python
import matplotlib.pyplot as plt

img = plt.imread('cat.jpg')
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.subplot(1, 2, 2)
plt.axis(False)
plt.imshow(img)
plt.show()
````

````python
# Type
print(type(img))

# Channels
print("Channels order: RGB")

# Write the image
plt.imsave('cat_PLT.jpg', img)
````

```plt.imshow``` is able to display numpy arrays and PIL images (no converting is needed). 

## OpenCV
This famous library comes with many image processing functionalities. However, we might not use it that often in this course.

````python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('cat.jpg')
# cv2.imshow(img) # Doesn't work on Colab
cv2_imshow(img)
````

````python
# Type
print(type(img))

# Channels
print("Channels order: BGR!!!")

# Write the image
cv2.imwrite('cat_CV2.jpg', img)
````

<font color='red'>Note: </font> Since since the order of channels in OpenCV is BGR, and matplotlib expcets RGB colors get messed up if we `plt.imshow`.

````python
plt.imshow(img)
plt.show()
````

````python
# Easy fix
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
plt.show()
````
