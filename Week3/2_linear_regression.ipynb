```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
```

```python
# Generate a random dataset
X = np.array([
       [ 0.04518035],
       [-0.84359252],
       [ 0.35966431],
       [-0.16797653],
       [-1.75281623],
       [-1.13996135],
       [-0.31996113],
       [-0.39066178],
       [-0.53555869],
       [-0.1028942 ],
       [-1.27873949],
       [-1.54783272],
       [-1.5066591 ],
       [ 0.08302513],
       [ 0.58614565],
       [ 1.46811296],
       [ 1.24930383],
       [-0.28339964],
       [ 0.22662137],
       [ 1.47788056]])

y = np.array([  2.49812544, -14.42378272,   3.74268996,  -4.02327442,
       -20.40151345, -14.91702401,  -2.467852  ,  -1.3403185 ,
        -7.27402561,  -4.46286526, -14.72417193, -18.46602318,
       -16.87112607,   1.96543249,   1.06066203,  19.76885427,
        20.93869911,  -1.77464117,   4.02257209,  18.04709569]).reshape(-1, 1)


# Visualize the dataset
plt.scatter(X[:, 0], y)
plt.title('Data-Points')
plt.xlabel('Feature1')
plt.ylabel('Feature2')

plt.xlim(X.min()-0.1, X.max()+0.1)
plt.ylim(y.min()-0.1, y.max()+0.1)
plt.grid()
plt.show()
```

```python
# Generate a random line
theta = np.random.randn(1,)
bias = np.ones((1,))
print(f"Line equation:\n{theta.item():.3f} X + {bias.item():.3f} = Y")
fn = lambda x, theta, bias: (theta[0]*x + bias)

# Visualize the line
line_domain = np.arange(-2, 2, 0.1)
plt.scatter(X[:, 0], y)
plt.plot(line_domain, fn(line_domain, theta, bias), 'r')
plt.scatter(X[:, 0], fn(X[:, 0], theta, bias), color='red')
plt.xlabel('Feature1')
plt.ylabel('Feature2')

plt.xlim(X.min()-0.1, X.max()+0.1)
plt.ylim(y.min()-0.1, y.max()+0.1)
plt.title('Random Line Predictions')
plt.grid()
plt.show()
```

```python
# Visualize the error
plt.title("Error of Prediction")
plt.scatter(X[:, 0], y)
plt.plot(line_domain, fn(line_domain, theta, bias), 'r')
plt.scatter(X[:, 0], fn(X[:, 0], theta, bias), color='red')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
for x_temp, y_temp in zip(X.reshape(-1), y):
    py_temp = fn(x_temp, theta, bias)
    plt.plot((x_temp, x_temp), (y_temp, py_temp), color='orange')


plt.xlim(X.min()-0.1, X.max()+0.1)
plt.ylim(y.min()-0.1, y.max()+0.1)
plt.grid()
plt.show()
```

```python
# Set the learning rate
eta = 0.001

verbose = 100

# Number of iterations
n_iterations = 1000

# Initialize the cost array
costs = []

for i in range(n_iterations):
    # Compute predictions
    y_pred = X * theta + bias

    # Compute the error
    diff = y_pred - y
    error = np.sum(np.square(diff))

    # Compute the gradient
    # Double check the gradient (I don't promise its correctness)
    theta_gradients = X.T @ (2 * diff)
    bias_gradients = np.sum(diff)
    # Update the parameters
    theta = theta - eta * theta_gradients
    bias = bias - eta * bias_gradients

    if i % verbose == 0:

        print(f"Line equation:\t{theta.item():.3f} X + {bias.item():.3f} = Y")
        print(f"Error: {error:.3f}")
        plt.scatter(X[:, 0], y)
        plt.plot(line_domain, fn(line_domain, theta, bias), 'r')

        plt.xlabel('Feature1')
        plt.ylabel('Feature2')

        plt.xlim(X.min()-0.1, X.max()+0.1)
        plt.ylim(y.min()-0.1, y.max()+0.1)
        plt.grid()
        plt.show()
```
