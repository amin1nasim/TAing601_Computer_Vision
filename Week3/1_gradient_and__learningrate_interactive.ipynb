```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
from IPython.display import clear_output
```

```python
######
# It doesn't work on colab
# Uncomment the line below if you are running locally
#%matplotlib notebook
######
def gradient(x):
    return 2*x
def move(event):
    global X
    learning_rate = widget_learning_rate.value/1000
    g = gradient(X[-1])
    X.append(X[-1] - learning_rate * g)
    print('X:({0:.2f},{1:.2f})  Learning Rate:{2:.3f}  Gradient:{3:.2f}  New X:({4:.2f},{5:.2f})'.format(
        X[-2], X[-2]**2, learning_rate, g, X[-1], X[-1]**2))
    plt.plot(t, t**2, 'b', X , np.square(X), 'r', X , np.square(X), 'rd')
    plt.show()

# Plot
t = np.arange(-200, 200.5, 1)
X=[-145,]
plt.plot(t, t**2, 'b',X , np.square(X), 'rd')
plt.show()

print("Learning Rate * 1e-3")
widget_learning_rate = widgets.FloatSlider(min=1, max=2000., value=1
                                      , description='LR:')
display(widget_learning_rate)
widget_next = widgets.Button(description='Move')
display(widget_next)
widget_next.on_click(move)
```
